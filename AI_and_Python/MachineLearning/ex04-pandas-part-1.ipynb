1 Read the data from the csv file and create a DataFrame from it.

```python
import pandas as pd

df = pd.read_csv("data_files/earthquakes.csv")
df
```

2 Remove the following columns from the DataFrame: *Magnitude Seismic Stations*, *Azimuthal Gap*, *Horizontal Distance* and *Horizontal Error*.

```python
df.drop(["Magnitude Seismic Stations", "Azimuthal Gap", "Horizontal Distance","Horizontal Error"], axis=1, inplace=True)
df
```

3 Filter data from the DataFrame so that only rows matching the following criteria will be shown:
    - Source must be US
    - Magnitude value must be over 7.8
    - ID value must start with the string "USP0007"

```python
display(df.loc[(df["Source"] == "US") & (df["Magnitude"] > 7.8) & (df["ID"].str.startswith("USP0007"))])
```

4 Create a new column to the tenth column position called Grade which will have values from the following formula: ("Magnitude" * "Depth") / "Root Mean Square". Results in this column should be presented with three decimals.

```python
#df.apply(lambda x: (x["Magnitude"] * x["Depth"])/x["Root Mean Square"].astype(np.float) ), axis=1)

#10th position
#3 decimals
# tämä toimii ilman indeksipaikkaa df["Grade"] = ((df["Magnitude"] * df["Depth"])/df["Root Mean Square"].astype(float)).round(3)
#df.drop(["Grade"], axis=1, inplace=True)

#tenth column position is index 9
df.insert(9, "Grade", ((df["Magnitude"] * df["Depth"])/df["Root Mean Square"].astype(float)).round(3) )

df
```
