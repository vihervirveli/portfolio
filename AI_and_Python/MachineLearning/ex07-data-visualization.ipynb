1 Read the earthquakes.csv file and save its content into a variable. Then use bar chart to visualize magnitude values being 8.2 or over so that date values will be presented in x-axle and magnitude in y-axle.

```python
import pandas as pd
df = pd.read_csv("data_files/earthquakes.csv")
df
```

```python
vis = df.loc[(df["Magnitude"] >= 8.2)].plot(kind="bar",x="Date", y="Magnitude",figsize=(6,5))
vis.set_xlabel("Date")
vis.set_ylabel("Magnitude")
```

2 Read the gamedata.json file to a new variable. Visualize the data with pie chart so that share count will be counted for each Steam rating (in other words how many percentage out of all games have received which rating).

```python
import pandas as pd

df1 = pd.read_json("data_files/gamedata.json")

df1
```

```python
df1["steamRatingText"] = df1["steamRatingText"].replace([None],"None")

pie = df1["steamRatingText"].copy().value_counts(normalize=True).rename_axis("Ratings").to_frame("Percentages")
pie["Percentages"] = pie["Percentages"].apply(lambda x: round(x * 100, 1))
pie
```

```python
pie["Ratings"] = pie.index
```

```python
label_list = [str(i) + ' %' for i in pie["Percentages"]]
dfplot = pie.plot(kind="pie",y="Percentages",labels=label_list,figsize=(10,5), title="How many percent of games received each kind of review?").legend(pie["Ratings"], bbox_to_anchor=(1.5, 1) )
```
